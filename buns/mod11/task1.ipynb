Без использования numpy:

```python
def mult(a, b):
    n = len(a)
    result = [[0] * n for _ in range(n)]

    for i in range(n):
        for j in range(n):
            for k in range(n):
                result[i][j] += a[i][k] * b[k][j]
    return result
```

C использованием numpy

```python
def np_mult(a, b):
    result = np.dot(a, b)
    return result
```

```python
%%time
# засечем время работы функции без NumPy
M1 = mult(a, b)
```

```python
%%time
# засечем время работы функции с NumPy
M2 = np_mult(a, b)
```

```python
# проверим корректность
np.abs(np.array(M1) - M2).sum()
```

Почему методы ***numpy*** оказываются эффективнее? Что вы можете сказать о скорости вычислений в NumPy?

**Ответ:**

Методы NumPy оказываются эффективнее по нескольким причинам. Во-первых, NumPy написан на языке программирования C, что позволяет значительно ускорить вычисления по сравнению с использованием чистого Python. Язык C является низкоуровневым и позволяет более эффективно использовать аппаратные возможности компьютера.

Во-вторых, NumPy использует векторизацию, что позволяет выполнять операции над массивами данных целиком, а не поэлементно. Векторизация позволяет использовать оптимизированные аппаратные инструкции для выполнения операций над массивами, ускоряя обработку данных. Например, при выполнении математических операций над массивами NumPy использует встроенные функции библиотеки BLAS (Basic Linear Algebra Subprograms), которые оптимизированы для работы с матрицами и векторами.

Кроме того, методы NumPy обеспечивают более эффективное использование памяти. Вместо создания новых объектов с каждой операцией, NumPy позволяет выполнять операции в place, т.е. непосредственно внутри исходного массива, что значительно снижает накладные расходы на создание и удаление объектов.

Скорость вычислений в NumPy зависит от нескольких факторов, таких как размеры данных, сложность операций и оптимизации аппаратной платформы. Однако общим правилом является то, что использование методов NumPy значительно ускоряет вычисления по сравнению с чистым Python. Векторизация операций и оптимизация использования памяти позволяют выполнить большие вычислительные задачи в несколько раз быстрее, чем с использованием стандартных средств Python.

### Задание 3
Напишите функцию, которая находит сумму четных диагональных элементов квадратной матрицы. Если таких элементов нет, то вывести `0`.

```python
# функция, решающая задачу с помощью NumPy
def np_diag_2k(a):
  arr = np.array(a)
  n = arr.shape[0]
  res = 0

  for i in range(n):
      if arr[i, i] % 2 == 0:
          res += arr[i, i]
  return res
```

```python
# зададим некоторую квадратную матрицу
a = np.random.randint(1, 10, size=(5, 5))
a
```

```python
%%time
# засечем время работы функции с NumPy
np_diag_2k(a)
```

### Задание 4
На вход дан двумерный массив $X$. Напишите функцию, которая для каждой строчки $x = (x_1, x_2, \ldots, x_n)$ массива $X$  строит строчку $s = (s_1, s_2, \ldots, s_n)$, где $s_k=\frac{x_1+...+x_k}{k}$, а затем выдаёт массив из построенных строчек. Используйте библиотеку numpy (вам поможет функция cumsum). Выходом функции должен быть двумерный массив той же формы, что и $X$.

```python
# функция, решающая задачу с помощью NumPy
def np_sec_av(*A):
    result = np.cumsum(A, axis=1)
    return result
```

```python
# зададим некоторую последовательность и проверим ее на ваших функциях.
# функция должна работать ~ в 50 раз быстрее, чем без Numpy
A = sps.uniform.rvs(size=10**3)
print(A)

%time S2 = np_sec_av(A)
```

### Задание 5
Написать функцию для кодирование массива (Run-length encoding). Все подряд повторения элементов функция сжимает в один элемент и считает количество повторений. Функция возвращает кортеж из двух векторов одинаковой длины. Первый содержит элементы, а второй — сколько раз их нужно повторить.

Пример: encode(np.array([1, 2, 2, 3, 3, 1, 1, 5, 5, 2, 3, 3])) = (np.array[1, 2, 3, 1, 5, 2, 3]), np.array[1, 2, 2, 2, 2, 1, 2])

```python
# функция, решающая задачу с помощью NumPy
def np_encode(a):
    elements = []
    counts = []

    current_element = a[0]
    current_count = 1

    for i in range(1, len(a)):
        if a[i] == current_element:
            current_count += 1
        else:
            elements.append(current_element)
            counts.append(current_count)
            current_element = a[i]
            current_count = 1

    elements.append(current_element)
    counts.append(current_count)

    result = np.array(elements), np.array(counts)
    return result
```

### Задание 6

Пусть задан некоторый одномерный массив $X$. Надо построить новый массив, где все элементы с нечетными индексами требуется заменить на число $a$ (если оно не указано, то на 1). Все элементы с четными индексами нужно возвести в куб. Затем записать элементы в обратном порядке относительно их позиций. Массив $X$ при этом должен остаться *без изменений*. В конце требуется слить массив X с преобразованным X и вывести в обратном порядке.  

Пример:  
X = [100, 200, 300, 400, 500] -> [100, a, 300, a, 500] -> [500^3, a, 300^3, a, 100^3] -> склеиваем -> [100, 200, 300, 400, 500, 500^3, a, 300^3, a, 100^3]

```python
# функция, решающая задачу с помощью NumPy
from copy import copy

def np_transformation(X, a=1):
    """
    param X: array
    """
    transformed_X = X.copy()

    transformed_X[1::2] = a

    transformed_X[::2] = transformed_X[::2] ** 3

    reversed_transformed_X = transformed_X[::-1]

    result = np.concatenate((X, reversed_transformed_X))

    return result
```

```python
X = np.array([i for i in range(1, 10, 2)])
print(X)
# функция должна работать ~ в 20 раз быстрее, чем без Numpy
%time S2 = np_transformation(X, 5)
print(np_transformation(X, 5))
```
